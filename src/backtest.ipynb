### 模組一: 五日線、布林軌道
- 進場:突破上軌並且隔天沒有回跌昨天漲幅過半
- 出場:收盤價破布林軌道上軌

```python
%reload_ext autoreload
%autoreload 2

# 使用範例
import pandas as pd
from lib.backtest.strategy_one import backtest_strategy



df1 = pd.read_csv("../data/TPE-sample1.csv", encoding="utf-8")
df1 = backtest_strategy(df1, ma_period=15, bb_period=20, bb_std=1.83, drop_threshold=0.72)
df1
```

```python
%reload_ext autoreload
%autoreload 2

from lib.performance_analysis import calculate_strategy_performance

# 從 position 欄位產生進出場訊號
# 當 position 從 0 變成 1 時，是進場訊號
df1['entry'] = (df1['position'] == 1) & (df1['position'].shift(1) == 0)

# 當 position 從 1 變成 0 時，是出場訊號
df1['exit'] = (df1['position'] == 0) & (df1['position'].shift(1) == 1)

# 計算策略績效
performance = calculate_strategy_performance(
    df1,
)

# 印出績效報告
performance_df = pd.DataFrame([performance])

# 存成csv檔案
# performance_df.T.to_csv('module_one.csv',
#                         header=['Value'],  # 欄位名稱 (y 軸)
#                         index_label='Metric')  # 列名 (x 軸)

performance_df.T
```
